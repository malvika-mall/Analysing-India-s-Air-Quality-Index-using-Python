# 05 Analyzing Proportion of missing values in each column with Bar chart.

```python
import missingno as msno
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#finding the proportion of missing values in each column
missing=pd.DataFrame(df_city.isna().sum()/len(df_city))
missing.columns=['Proportion']
print(missing.sort_values(by='Proportion', ascending=False))

#plotting the number of non-null values in each column
msno.bar(df_city)
```

# As the above bar graph clearly indicates that PM10 and Xylene have high proportion of missing values. This might be due to manual inconsistencies or human negligence while recording or absence of enough stations in the city.

```python
#filling missing values with zero 
df_city.fillna(0,inplace=True)
```

# The missing values can also be imputed by way of mean but i choose to replace it with zero to avoid skewing of data.

```python
#finding the proportion of missing values in each column stations
import missingno as msno
missing=pd.DataFrame(df_stations.isna().sum()/len(df_stations))
missing.columns=['Proportion']
print(missing.sort_values(by='Proportion', ascending=False))

#plotting the number of non-null values in each column
msno.bar(df_stations)
```

```python
#filling missing values with zero 
df_stations['AQI'].fillna(0,inplace=True)
df_stations['AQI_Bucket'].fillna('Not Provided',inplace = True)
```

```python
#extracting year and month for each record
df_city['year'] = pd.DatetimeIndex(df_city['Date']).year
df_city['month'] = pd.DatetimeIndex(df_city['Date']).month
```

```python
#subsetting data in relevan group
df_city['PM']=df_city['PM2.5'] + df_city['PM10']
```

```python
#subsettinubbing nitrogen oxides
df_city['Nitric']=df_city['NO'] + df_city['NO2']+ df_city['NOx']
```

```python
# subsetting Benzene, toluene and Xylene together
df_city['BTX']=df_city['Benzene'] + df_city['Toluene']+ df_city['Xylene']
```

```python
#grouping pollutant levels in every city by year and month
cities_group_ym=df_city.groupby(['State','City','year','month'])[['PM','Nitric','CO','NH3','O3','SO2','BTX','AQI']].mean()

cities_group_ym=cities_group_ym.reset_index(['City','year','month'])
cities_group_ym.head()
```

```python
# Exporting prepared data
cities_group_ym.to_csv(os.path.join(path, '02 Data','Prepared Data', 'city_checked.csv'))
```
